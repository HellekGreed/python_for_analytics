```python
df = pd.read_csv("kc_house_data.csv", encoding="ISO-8859-1")
df.head()
```

### Задание 1.1

Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади

```python
df["price_per_sq_lot"] = df["price"] / df["sqft_living"]
df.head()
```

### Задание 1.2

Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.
Если реновации дома не было, то в новом признаке поставьте 0

```python
df["delta_renovated"] = (df["yr_renovated"] - df["yr_built"])
df["delta_renovated"] = df["delta_renovated"].where(df["delta_renovated"] > 0, 0)
df.head()
```

### Задание 1.3

Создайте признаки года продажи, месяца продажи

```python
df["date"] = pd.to_datetime(df["date"], format="%Y%m%dT000000")
df["year_sale"] = df["date"].dt.year
df["month_sale"] = df["date"].dt.month
df.head()
```

### Задание 1.4

Удалите признаки date, zipcode, lat, long

```python
df.drop(["date", "zipcode", "lat", "long"], axis=1, inplace=True)
df.head()
```

## Задание 3

Составьте несколько сводных таблиц

### Задание 3.1

Найдите среднюю стоимость домов в зависимости от количества спален.
Отсортируйте от меньшей стоимости к большей.

```python
df.groupby("bedrooms").agg({"price": "mean"}).sort_values(by="price").round(2)
```

### Задание 3.2

Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома.

```python
df.groupby("condition").agg({'price': ['min', 'mean', 'max']}).round(2)
```

### Задание 3.3

Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида.

```python
df.pivot_table(index='view',
               columns='waterfront',
               values='id',
               aggfunc='count',
               fill_value=0)
```

### Задание 3.4

Каких домов в зависимости от этажности и количества спален больше?

```python
df.groupby(["floors", "bedrooms"]).agg(count=('id', "count")).sort_values("count", ascending=False).head(1)
```

### Задание 3.5

Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома.

```python
df.pivot_table(index='condition',
               columns='grade',
               values='price',
               aggfunc='median',
               fill_value=0)
```
